```python
import pandas as pd

d1 = pd.read_parquet('./FinalDataset/Device1.parquet')
d2 = pd.read_parquet('./FinalDataset/Device2.parquet')
d3 = pd.read_parquet('./FinalDataset/Device3.parquet')
d4 = pd.read_parquet('./FinalDataset/Device4.parquet')
d5 = pd.read_parquet('./FinalDataset/Device5.parquet')
d6 = pd.read_parquet('./FinalDataset/Device6.parquet')
d7 = pd.read_parquet('./FinalDataset/Device7.parquet')
d8 = pd.read_parquet('./FinalDataset/Device8.parquet')
d9 = pd.read_parquet('./FinalDataset/Device9.parquet')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Combine all the dataframes into a single dataframe
df = pd.concat([d1, d2, d3, d4, d5, d6, d7, d8, d9])

# Define the labels
labels = {
    'benign': 'Benign',
    'mirai_ack': 'Mirai ACK',
    'mirai_scan': 'Mirai Scan',
    'mirai_syn': 'Mirai SYN',
    'mirai_udp': 'Mirai UDP',
    'mirai_udpplain': 'Mirai UDP Plain',
    'gafgyt_combo': 'Gafgyt Combo',
    'gafgyt_scan': 'Gafgyt Scan',
    'gafgyt_tcp': 'Gafgyt TCP',
    'gafgyt_udp': 'Gafgyt UDP',
    'gafgyt_junk': 'Gafgyt Junk'
}

# Count the occurrences of each label
label_counts = df['type'].value_counts()

# Separate the benign and anomalous data
benign_count = label_counts['benign']
anomalous_count = label_counts.sum() - benign_count

# Set plot style
sns.set(style='whitegrid')
colors = sns.color_palette('Set2')

# Create a figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Create a bar plot
sns.barplot(x=['Benign', 'Anomalous'], y=[benign_count, anomalous_count], palette=[colors[0], colors[1]], ax=ax)

# Add data labels to the bars
for i, v in enumerate([benign_count, anomalous_count]):
    ax.text(i, v + 1000, str(v), ha='center', va='top', fontsize=12, weight='bold')

# Set axis labels and title
ax.set_xlabel('Data Type', fontsize=14, labelpad=12)
ax.set_ylabel('Count', fontsize=14, labelpad=12)
ax.set_title('Distribution of Benign Data and Anomalous Data among Nine Devices', fontsize=16, weight='bold')

# Customize the tick labels
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Adjust the plot layout
plt.tight_layout()

# Show the plot
plt.show()
```
